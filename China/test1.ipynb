```python
import akshare as ak
import pandas as pd
import datetime
```

```python
# Get the list of industry names
stock_board_industry_name_em_arr = ak.stock_board_industry_name_em()["板块名称"]
# Extract all stocks from the industry board
industry_stocks_df = pd.DataFrame(columns=["行业", "代码", "名称"])
for industry_name in stock_board_industry_name_em_arr:
    stock_board_industry_cons_em_df = ak.stock_board_industry_cons_em(symbol=industry_name)
    for row in stock_board_industry_cons_em_df.itertuples():
        industry_stocks_df.loc[len(industry_stocks_df)] = [
            industry_name,
            row.代码,
            row.名称
        ]
```

```python
import pandas as pd

df = pd.read_csv("reports/股票筛选报告-raw-20250711.csv", index_col=0)
```

```python
days = 29
# 总市值 < 200 亿, 动态市盈率 < 50, {days}日主力净流入-总净额(亿) > 1, {days}日涨跌幅(%) < 10
df = df[(df[f"{days}日主力净流入-总净额(亿)"] > 1) & (df[f"{days}日涨跌幅(%)"] < 10)]
df
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df.sort_values(by=["行业", f"{days}日平均涨跌幅(%)"], inplace=True)
df.reset_index(inplace=True, drop=True)
df.to_csv(f"reports/股票筛选报告-20250711.csv", index=True)
```
